### 单变量线性回归

```python
data = np.loadtxt('../../Demo_Ref/Linear-models/linear_regression/linear_regression_data1.txt', delimiter=',')

X = np.c_[np.ones(data.shape[0]),data[:,0]]
y = np.c_[data[:,1]]

```

```python
plt.scatter(X[:,1], y, s=30, c='r', marker='x', linewidths=1)
plt.xlim(4,24)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s');
```

```python
# 计算损失函数
def computeCost(X, y, theta=[[0],[0]]):
    m = y.size
    J = 0
    
    h = X.dot(theta)
    
    J = 1.0/(2*m)*(np.sum(np.square(h-y)))
    
    return J
```

```python
computeCost(X,y)
```

```python
#梯度下降（可以优化，迭代的过程中选择性的将某些权重该次迭代不参与运算）
def gradientDescent(X, y, theta=[[0],[0]], alpha=0.01, num_iters=1500):
    m = y.size
    J_history = np.zeros(num_iters)
    
    for iter in np.arange(num_iters):
        h = X.dot(theta)
        theta = theta - alpha*(1.0/m)*(X.T.dot(h-y))
        J_history[iter] = computeCost(X, y, theta)
    return(theta, J_history)

#生成num个0-ranges之间的随机数
def Rnd(ranges,num):
    L = [random.randint(0,ranges) for _ in range(num) ]
    return L

def gradientDescent_opt(X, y, theta=[[0],[0]], alpha=0.01, num_iters=1500,num_batch_scale=0.75):
    m = y.size
    num = int(m * num_batch_scale)
    J_history = np.zeros(num_iters)
    
    for iter in np.arange(num_iters):
        rand_arr = np.arange(num)
        np.random.shuffle(rand_arr)
        X_opt = X[rand_arr[0:num]]
        y_opt = y[rand_arr[0:num]]
        h = X_opt.dot(theta)
        theta = theta - alpha*(1.0/num)*(X_opt.T.dot(h-y_opt))
        J_history[iter] = computeCost(X_opt, y_opt, theta)
    return(theta, J_history)
```

```python
# 画出每一次迭代和损失函数变化
theta , Cost_J = gradientDescent(X, y)
print('theta: ',theta.ravel())

plt.plot(Cost_J)
plt.ylabel('Cost J')
plt.xlabel('Iterations');
```

```python
# 画出每一次迭代和损失函数变化
theta , Cost_J = gradientDescent_opt(X, y,[[0],[0]],0.01,1500,1.0)
print('theta: ',theta.ravel())

plt.plot(Cost_J)
plt.ylabel('Cost J')
plt.xlabel('Iterations');
```

```python
xx = np.arange(5,23)
yy = theta[0]+theta[1]*xx

# 画出我们自己写的线性回归梯度下降收敛的情况
plt.scatter(X[:,1], y, s=30, c='r', marker='x', linewidths=1)
plt.plot(xx,yy, label='Linear regression (Gradient descent)')

# 和Scikit-learn中的线性回归对比一下 
regr = LinearRegression()
regr.fit(X[:,1].reshape(-1,1), y.ravel())
plt.plot(xx, regr.intercept_+regr.coef_*xx, label='Linear regression (Scikit-learn GLM)')

plt.xlim(4,24)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.legend(loc=4);
```
